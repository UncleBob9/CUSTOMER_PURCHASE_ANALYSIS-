```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\HP\Desktop\project imar\customer data\retail_store_sales1.csv")
df.head(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["Item"]=df["Item"].fillna("Unknown Item")
```

```python
df["Price Per Unit"]=df["Total Spent"]/df["Quantity"]
```

```python
df.loc[
    df["Price Per Unit"].isna(),
    ["Total Spent", "Quantity"]
].isna().sum()
```

```python
df=df.dropna(subset=["Total Spent","Quantity","Price Per Unit"])
```

```python
df.loc[:,"Discount Applied"]=df["Discount Applied"].fillna("False")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["Transaction ID"].duplicated().sum()
```

```python
df.dtypes
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors='coerce')
```

```python
df["Year"]=df["Transaction Date"].dt.year
df["Month"]=df["Transaction Date"].dt.month
df["Day"]=df["Transaction Date"].dt.day
df["Weekday"]=df["Transaction Date"].dt.day_name()
```

```python
df.info()
```

```python
df.head()
```

```python
(df["Quantity"] % 1 !=0).sum()
```

```python
df.loc[:,"Quantity"]=df["Quantity"].astype(int)
```

```python
Q1=df["Total Spent"].quantile(0.25)
Q3=df["Total Spent"].quantile(0.75)
IQR=Q3-Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers=(df["Total Spent"]<lower_bound)| (df["Total Spent"]>upper_bound)
outlier_rows=df[outliers]
len(outlier_rows)
```

```python
invalid_rows=df[
    (df["Price Per Unit"] <= 0)|
    (df["Quantity"] <= 0)|
    (df["Total Spent"] <= 0)
]
```

```python
invalid_rows
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df["Total Spent"])
plt.show()
```

```python
df_cleaned = df[~outliers]
```
